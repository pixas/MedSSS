```python
from Evol_Instruct import client 
import os 
import pandas as pd 
import numpy as np
import json
from Evol_Instruct.utils.eval_method import compute_dataset_with_method
```

```python
import subprocess
def show_comparison_with_selection(log_dir, save_name):
    datasets = ['MedQA', 'MedMCQA', 'pubmedqa', 'bioasq', 'med_mmlu', 'healthfact', 'ddxplus', 'pubhealth', 'drugdose', 'seer', 'biomrc']
    
    results = {}
    settings = ['prm-minmax', 'prm-max', 'prm-min-vote-sum', 'prm-vote-sum', 'vote-sc']
    
    
    results['settings'] = settings
    for dataset in datasets:
        results[dataset] =[]
        n = 16
        merge_file = os.path.join(log_dir, dataset, save_name)
        if not os.path.exists(merge_file):
            results[dataset].extend([np.nan] * len(settings))
            continue
        try:
            client.read(merge_file)
        except:
            print(merge_file)
            results[dataset].extend([np.nan] * len(settings))
            continue
        acc_list, cache = compute_dataset_with_method(merge_file, settings)
        results[dataset].extend(list(x * 100 for x in acc_list.values()))
       
    df = pd.DataFrame(results)
    df['Mean'] = df.apply(lambda x: x[1:].mean(), axis=1)
    return df
```

```python
log_dir = 'the log file directory'
save_name = 'the log file name which should end of .jsonl'
setting_df = show_comparison_with_selection(log_dir, save_name)
setting_df
```
